````python
# iris dataset の読み込み（第2回課題のネタバレ）
from sklearn.datasets import load_iris
iris_data = load_iris() #del iris_data などで不要となった変数を消すことも可能
data, label, feature_names = iris_data["data"], iris_data["target"], iris_data["feature_names"]
iris_df = pd.DataFrame(data, columns=feature_names)
label = pd.DataFrame(label, columns=["label"])
iris_df = pd.concat([iris_df, label], axis=1)
iris_df.head()
````

## hist

### 課題1
sepal length (cm) について，histogram を作る（bins をいじるなどしてよい）．
```
plt.hist
```

````python
x = iris_df['sepal length (cm)']
n, bins, patches = plt.hist(x, 20)
````

##### 課題2
課題1の graph に label を付け，判例を付与．さらに，title を fontsize 18 で追加（できれば xlabel, ylabel もやってほしいが）．
```
#plt.figure
plt.hist
#plt.xlabel
#plt.ylabel
plt.title
plt.legend
```

````python
#plt.figure
plt.hist(x,20,label="sepal length (cm)")
plt.xlabel("sepal length (cm)")
plt.ylabel("sum")
plt.title("sepal length (cm) histgram" ,size=18)
plt.legend()
````

### 発展課題１
fontsize を backend([rcParams](https://matplotlib.org/3.2.1/api/matplotlib_configuration_api.html#matplotlib.RcParams))で設定
```
plt.rcParams
#plt.figure
plt.hist
#plt.xlabel
#plt.ylabel
plt.title
plt.legend
```

````python
plt.rcParams['font.size']=18
#plt.figure
plt.hist(x,20,label="sepal length (cm)")
plt.xlabel("sepal length (cm)")
plt.ylabel("sum")
plt.title("sepal length (cm) histgram")
plt.legend()
````

### 発展課題2
1行2列で可視化しよう（subplot の使用）
```
plt.rcParams
#plt.figure
plt.hist
#plt.xlabel
#plt.ylabel
plt.title
plt.legend
```

````python
#何を可視化したらいいのかいまいち分からず，とりあえずsepal lengthとpetal lengthの2つを描画しました

plt.rcParams['font.size']=18
plt.rcParams["figure.figsize"] = (12.8, 4.8)

ax1=plt.subplot(121)
plt.hist(x,20,label="sepal length (cm)")
plt.xlabel("sepal length (cm)")
plt.ylabel("sum")
plt.title("sepal length (cm)")
plt.legend()

y=iris_df['petal length (cm)']
ax2=plt.subplot(122)
plt.hist(y,20,label="petal length (cm)")
plt.xlabel("petal length (cm)")
plt.ylabel("sum")
plt.title("petal length (cm)")
plt.legend()

#課題3以降のグラフの描画でグラフのサイズに影響が出たのでデフォルトに戻しました
plt.rcParams["figure.figsize"] = (6.4, 4.8)
````

## scatter

### 課題3
scatter を用いて，横を sepal width，縦を sepal length とする点を表示
```
plt.scatter
```

````python
x ,y = iris_df["sepal width (cm)"], iris_df["sepal length (cm)"]
plt.scatter(x,y)
````

### 課題4
課題3の図に，横を petal width，縦を petal length とする緑色の点を追加で表示
```
plt.scatter
plt.scatter
```

````python
x ,y = iris_df["sepal width (cm)"], iris_df["sepal length (cm)"]
x_i ,y_i = iris_df["petal width (cm)"], iris_df["petal length (cm)"]
plt.scatter(x,y,c="b")
plt.scatter(x_i,y_i,c="g")
````
